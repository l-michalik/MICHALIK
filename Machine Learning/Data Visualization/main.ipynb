```python
import pandas as pd
data = pd.read_csv('../data/Laptop_price.csv')
```

```python
%pip install matplotlib
import matplotlib.pyplot as plt
```

```python
data.groupby('Brand')['Price'].mean().plot(kind='bar')
plt.title('Avg Price by Brand')
plt.xlabel('Brand')
plt.show()
```

```python
data['Brand'].value_counts().plot(kind='pie')
plt.ylabel('')
```
